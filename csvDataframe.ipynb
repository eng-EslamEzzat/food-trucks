```python
import pandas as pd
```

```python
df = pd.read_csv("food-truck-data.csv")
df.head(2)
```

```python
from math import dist
def get_dist(lat1, long1, lat2, long2): 
    return dist((lat1, long1), (lat2, long2))
```

```python
test_lat_long = (37.76008693198698, -122.41880648110114)
all_dists = []
for id, lat, long in zip(df['locationid'],df['Latitude'], df['Longitude']):
    all_dists.append((id, get_dist(*test_lat_long, lat, long)))
all_dists.sort(key=lambda x:x[1])
```

```python
all_dists[:5]
a, b = zip(*all_dists[:5])
```

```python
newdf = df[df['locationid'].isin(a)].copy()
newdf
```

```python
newdf.to_json(orient='records')
```
